```python
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR

#import urllib.request

auth_data = {
    'grant_type'    : 'client_credentials',
    'client_id'     : '72402d69cb754a1fada1572d49a56afa',
    'client_secret' : '932584ccebaf9ab0ea349b83736f6201cc70258abe25be77b127bbef585ff96a',
    'scope'         : 'read_product_data read_financial_data read_content'
}
# create Session instance
session = requests.Session()

# make a POST to retrieve access_token
auth_request = session.post('https://idfs.gs.com/as/token.oauth2', data = auth_data)
access_token_dict = json.loads(auth_request.text)
access_token = access_token_dict['access_token']

# update session headers
session.headers.update({'Authorization':'Bearer '+ access_token})

print(access_token)
```

```python
def getScores(symbol):
    payload = {
        "startDate": "2012-11-02",
        "endDate": "2017-11-04",
        "where": {
            "ticker": [ symbol ]
        }
    }

    request_url = 'https://api.marquee.gs.com/v1/data/USCANFPP_MINI/query'
    request = session.post(url=request_url, json = payload)
    results = json.loads(request.text)
    data = results['data']
    inteS=[]
    date=[]
    rs=[]
    growth=[]
    multiple=[]
    for i in range(len(data)):
        inteS.append((data[i]['integratedScore']))
        rs.append((data[i]['financialReturnsScore']))
        growth.append(data[i]['growthScore'])
        multiple.append(data[i]['multipleScore'])
        spli=data[i]['date'].split('-')
        date.append(spli[0]+spli[1]+spli[2])

    d = {'date': date, 'integrated score': inteS, 'finantial return score': rs, 'growthScore': growth, 'multipleScore':multiple}
    df = pd.DataFrame(data=d)
    return df

# plt.figure(0)
# plt.plot(date,inteS)
# plt.title("integrated score")

# plt.figure(1)
# plt.plot(date,rs)
# plt.title("finantial return score")

# plt.figure(2)
# plt.plot(date,growth)
# plt.title("growth score")
# plt.figure(3)
# plt.plot(date,multiple)
# plt.title("multiple score")
# plt.show()
```
